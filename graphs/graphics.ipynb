# Imports

```python
import matplotlib.pyplot as plt
```

# Variables

```python
# Sample data used from notes/sync_time_records.md
numbers = [88.2421282, 33.3676, 4.4086958, 2.7961284]
labels = ["Python","Java","C-Sharp","C"]

# Create a figure and axis
plt.figure(figsize=(10, 6))

# Create bar plot
bars = plt.bar(labels, numbers, color='skyblue', edgecolor='navy')

# Add a title and labels
plt.title('Time of Counting to a Billion Across Various Languages', fontsize=16)
plt.xlabel('Programming Languages', fontsize=12)
plt.ylabel('Execution Time (Seconds)', fontsize=12)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'{height}', ha='center', va='bottom')

# Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
